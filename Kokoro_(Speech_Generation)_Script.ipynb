```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch


from IPython.display import HTML, display
import base64
```

```python
pipeline = KPipeline(lang_code='a')


text = '''Welcome to the future of generative AI. Today, you will see how machines can talk, learn, and think creatively.'''


generator = pipeline(
    text, voice='am_puck', # <= change voice here
    speed=1, split_pattern=r'\n+'
)

speech = next(generator)
display(Audio(data=speech[2], rate=24000))
sf.write(f'audio.wav', speech[2], 24000) # save each audio file
```
